```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 2*np.pi
w_0 = 1.5*w
b = 0.25*w_0

L = 1.077

dt = 0.001
T = 500

x0 = -np.pi/2
dx0 = 0

def sim(x0, dx0, w, w_0, b, L, dt, T):
    x = [x0]
    dx = [dx0]
    t = [0]
    d2x = [L*w_0**2*np.cos(w*t[-1])-2*b*dx[-1]-w_0**2*np.sin(x[-1])]
    
    while t[-1] < T:
        t += [t[-1]+dt]
        dx += [dx[-1]+d2x[-1]*dt]
        x += [x[-1]+dx[-1]*dt]
        d2x += [L*w_0**2*np.cos(w*t[-1])-2*b*dx[-1]-w_0**2*np.sin(x[-1])]
    
    return(x, dx, t)
```

```python
x,dx, t = sim(x0, dx0, w, w_0, b, L, dt, T)
```

```python
plt.plot(t[0:10000], x[0:10000])
```

```python
L_list = np.linspace(1.06, 1.09, 500)

for L in L_list:
    x, dx, t = sim(x0, dx0, w, w_0, b, L, dt, T)
    plt.plot(np.linspace(L, L , len(x[-50000:-1][::1000])), x[-50000:-1][::1000], ".", color = "b", markersize=1)

plt.ylim(-0.6, 0)
```
